# 데이터 변환


```python
df['Date'] = pd.to_datetime(df['Date'],dayfirst = True)
type(df['Date'][0])
```

```python
df['Season'] = df['Season'].str.replace('Dec-13','12-13')
df['Season']
```

```python
df['Playing_Position'].fillna(method='ffill', inplace=True)
```

```python
df['Playing_Position'] = df['Playing_Position'].str.strip()
```

```python
def check_win(row):
    raw_result = ''.join([c for c in row['Result'] if c.isdigit() or c == ':'])
    home_goals, away_goals = map(int, raw_result.split(':')) #홈 골과 원정경기 골 구분

    outcome = 'draw' if home_goals == away_goals else 'win' if (home_goals > away_goals and row['Venue'] == 'H') \
    else 'win' if (home_goals < away_goals and row['Venue'] == 'A') else 'loss' if (home_goals < away_goals and row['Venue'] == 'H') \
    else 'loss'

    return outcome
```

```python
df['winLoss'] = df.apply(check_win, axis=1)
df[['Result','Venue', 'winLoss']].sample(3)
```

```python
goals_per_club = df.groupby('Club').size().reset_index(name='Goals').sort_values(by = "Goals", ascending=False)

color_map ={'Real Madrid':'#4D2DB7', 'Juventus FC':'#040D12', 'Manchester United':'#C70039', 'Al-Nassr FC':'#F8DE22','Sporting CP': 'green'}

fig = px.bar(goals_per_club, x='Club', y='Goals', title='Ronaldo Goals per Club',
             labels={'Club': 'Football Club', 'Goals': 'Number of Goals'}, text= 'Goals',
              color_discrete_map = color_map, color = 'Club')
fig.show()
```

```python
#호날두의 소속 클럽에서의 골 비율
fig = px.pie(goals_per_club,
             names='Club',
             values='Goals',
             title="Ronaldo's goals percentage per club %⚽%",
             labels={'Club':'Club', 'Goals':'Number of Goals'},
             color_discrete_map= color_map, color = 'Club')

custom_label = goals_per_club.apply(lambda row: f"{row['Club']}<br>{row['Goals']}", axis=1)
fig.update_traces(textinfo="label+percent", text=custom_label)
fig.show()
```

# 분석 결과


1. 클럽에서의 활약상

레알 마드리드와 멘체스터 유나이티드에서 비슷한 시즌(8시즌~9시즌)사이를 보냈지만, 호날두의 대부분의 골이 레알 마드리드 시절의 골이다. 가장 골을 많이 넣은 시즌들도 다 레알 마드리드 시절이고, 이때 전무후무한 기록인 챔피언스 리그 3연패를 했다.

마찬가지로, 유벤투스에서는 3시즌을 소화했지만, 맨유 시절과 골 차이가 적은 것으로 보아

각 클럽에서의 활약 상은
'레알 마드리드 > 유벤투스 > 맨체스터 유나이티드' 이라고 생각한다.

2. 득점 부위

원형 그래프의 결과에 따르면, 오른발의 비중이 높고, 그 다음으로는 페널티킥, 그 다음으로 왼발 슛, 헤더 순이다.

왼쪽 윙어로 뛴 경기가 많았고, 보통 왼쪽 사이드에서 골을 넣기 위해서는 감아 차거나 가까운 골대로 강한 슛팅을 해야하기 때문에, 오른발의 비중이 가장 높은 것 같다.


3. GAME CHANGER

꾸준한 득점력, 양쪽 사이드 윙 및 스트라이커 역할 소화에서 오는 지능적인 움직임, 훌륭한 조력자들과 함께 축구라는 게임을 자신의 놀이터로 만들었다고 본다.

관련 언사로

"내게 주입된, 진형을 형성하고 같은 팀의 공격수와 같이 따라가는 것에 관해 내가 배운 모든 미리 생각하는 전술적 이론들, 그것들이 2년 사이에 무용지물이 됐는데 왜냐하면 우리 팀에는 동료들의 도움을 받아 자기만의 규칙을 만들 수 있는 선수가 있었기 때문이다. 호날두는 축구를 재정의하는 것에 도움을 줬다."

라는 말처럼 그는 게임을 바꾸었다. 특히 리그보다 경기 수는 적지만, 그 중요도가 매우 큰 유럽대항전인 챔피언스리그에서 혼자 3골(혹은 그 이상)을 넣는 헤트트릭을 8번이나 할 정도로 큰 경기에서 유독 강했다. 현재도 그의 기록인 '챔피언스 최다 골'은 깨지지 않는 기록이다



호날두의 데이터를 분석한 결과는 그가 세계 최고의 축구 선수 반열에 드는 선수임에는 틀림없다고 생각한다.
